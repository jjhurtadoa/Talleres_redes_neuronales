```python
import math
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#
# A continuaci贸n se genera una muestra de 200 datos, con
# cuatro variables independientes, pero solo una explicativa
#
from sklearn.datasets import make_regression

np.random.seed(12345)

X, d = make_regression(
    n_samples=200,
    n_features=4,
    n_informative=1,
    bias=0.0,  # intercepto
    effective_rank=None,  # numero de vectores singulares para explicar los datos
    tail_strength=0.9,  # ? importancia relativa de las colas
    noise=12.0,
    shuffle=False,  # ?
    coef=False,  # ?coeficientes de la regresion
    random_state=12136,
)

#
# Grafica de la variable dependiente contra
# la variable explicativa
#
plt.scatter(X[:, 0], d)
plt.show()

#
# A continuaci贸n se estima el modelo de regresi贸n lineal
# con las 4 variables independientes
#
import statsmodels.api as sm

model = sm.OLS(d, X)
results = model.fit()
print(results.summary())
```

## Forward

```python
#
# Modelo que titne una sola variable X1
#
model = sm.OLS(d, X[:, 0])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que titne una sola variable X2
#
model = sm.OLS(d, X[:, 1])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que titne una sola variable X3
#
model = sm.OLS(d, X[:, 2])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que titne una sola variable X4
#
model = sm.OLS(d, X[:, 3])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene dos variables X1 y X2 
#
model = sm.OLS(d, X[:,[0,1]])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene dos variables X1 y X3
#
model = sm.OLS(d, X[:,[0,2]])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene dos variables X1 y X4
#
model = sm.OLS(d, X[:,[0,3]])
results = model.fit()
print(results.summary())
```

El mejor AIC se obtiene con X0, al agregar variables no se obtiene mejora entonces el proceso se detiene

##Backward

```python
#
# Modelo que tiene todas las variables
#
model = sm.OLS(d, X)
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene todas las variables menos X0
#
model = sm.OLS(d, X[:, [1,2,3]])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene todas las variables menos X1
#
model = sm.OLS(d, X[:, [0,2,3]])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene todas las variables menos X2
#
model = sm.OLS(d, X[:, [0,1,3]])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene todas las variables menos X3
#
model = sm.OLS(d, X[:, [0,1,2]])
results = model.fit()
print(results.summary())
```

Eliminando dos variables obtenemos el menor AIC, podemos escoger X1 o X3, en este caso escogeremos X3, ya haremos el mismo proceso. 

```python
#
# Modelo que tiene X0, X1 
#
model = sm.OLS(d, X[:, [0,1]])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene X0, X2 
#
model = sm.OLS(d, X[:, [0,2]])
results = model.fit()
print(results.summary())
```

```python
#
# Modelo que tiene X1, X2 
#
model = sm.OLS(d, X[:, [1,2]])
results = model.fit()
print(results.summary())
```
