```python
import math
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def f(x, y):
    return 100 * (x ** 2 - y) ** 2 + (1 - x) ** 2

X = np.arange(-2.048, 2.048, 0.025)
Y = np.arange(-0.1, 4.0, 0.025)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

def plot_contour():
    fig, ax = plt.subplots(figsize=(7, 7))
    ax.set_aspect("equal", "box")
    ax.contour(X, Y, Z)
    ax.grid()
    plt.plot([1], [1], "o", color="red")

def g(x, y):
    gx = 400 * x * (x ** 2 - y) - 2 * (1 - x)
    gy = -200 * (x ** 2 - y)
    return (gx, gy)

def improve(x, y, mu):
    gx, gy = g(x, y)
    x = x - mu * gx
    y = y - mu * gy
    return (x, y)

#
# Punto de inicio
#
x = -0.5
y = +3.5

#
#  almacena la historia de los puntos recorridos
#
history_x = [x]
history_y = [y]
history_f = [f(x, y)]

for epoch in range(11000):
    x, y = improve(x, y, 0.001)
    history_x.append(x)
    history_y.append(y)
    history_f.append(f(x, y))

plot_contour()
plt.plot(history_x, history_y, ".-", color="blue")
print(x, y)
print(f(x, y))
```
